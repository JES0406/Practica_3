```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.graph.Graph()
G.add_node(1)
G.add_node(2)

# We print the nodes of the graph
print(G)
print(G.nodes())

# Create a figure of 6 graphs
plt.figure(1, figsize=(12, 12))

# Create a graph in the figure 1
plt.subplot(321, title='Grafo sin aristas')
nx.draw(G, with_labels=True)

# Create a graph in the figure 2
G.add_edge(1, 2)
plt.subplot(322, title='Grafo con aristas')
nx.draw(G, with_labels=True)

# We add new nodes with the method add_nodes_from
G.add_nodes_from([3, 4, 5])
plt.subplot(323, title='Grafo con nodos aadidos en grupo')
nx.draw(G, with_labels=True)

# We add new edges with the method add_edges_from
G.add_edges_from([(1, 2), (1, 4), (3, 4), (2, 4), (3, 2), (3, 5)])
plt.subplot(324, title='Grafo tras aadir las aristas en grupo')
nx.draw(G, with_labels=True)

# We remove nodes with the method remove_node
G.remove_node(5)
plt.subplot(325, title='Grafo tras eliminar nodo 5')
nx.draw(G, with_labels=True)

# We now draw it with order
plt.subplot(326, title='Grafo con orden')
nx.draw(G, with_labels=True, pos={1:(0,1), 2:(1,1), 3:(0,0), 4:(1,0)})
```

```python
G2 = nx.graph.Graph()
nodos = ["a", "b", 10, 11, 3.14]
G2.add_nodes_from(nodos)
G2.add_weighted_edges_from([('a',10, 1.1), ("a", 3.14, 1.2), ("b", 11, 5.7)])

plt.figure(2, figsize=(12, 12))
plt.subplot(221, title='Grafo con nodos de distinto tipo y aristas con peso')
nx.draw(G2, with_labels=True)

# Para aadir contenido a las aristas
G3 = nx.graph.Graph()
G3.add_nodes_from(nodos)
G3.add_edges_from([('a',10, {'weight': 1.1, "object": "Una cadena"}), ("a", 3.14, {'weight': 1.2, "object": "Otra cadena"}), ("b", 11, {'weight': 5.7, "object": "Un objeto"})])

plt.subplot(222, title='Grafo con nodos de distinto tipo y aristas con peso y contenido')
nx.draw(G3, with_labels=True)
```

```python
Gd = nx.DiGraph()
nodos = [1, 2, 3, 4]
Gd.add_nodes_from(nodos)
Gd.add_edges_from([(1, 2),(2, 1),(1, 4),(3, 4),(2, 4),(4, 2),(3, 2)])

plt.figure(3, figsize=(12, 12))
plt.subplot(223, title='Grafo dirigido')
nx.draw(Gd, with_labels=True)
```

```python
print(f"Grados de G: {G.degree()}")
print(f"Grados de Gd: {Gd.degree()}")
print(f"Grados salientes de Gd: {Gd.out_degree()}")
print(f"Grados entrantes de Gd: {Gd.in_degree()}")
```

```python
print(f"Componentes conexas de G2: {list(nx.connected_components(G2))}")
print(f"Componentes conexas de Gd: {list(nx.weakly_connected_components(Gd))}")
print(f"Componentes fuertemente conexas de Gd: {list(nx.strongly_connected_components(Gd))}")
```

```python
tree = nx.bfs_tree(G, 1)
plt.figure(4, figsize=(12, 12))
plt.subplot(224, title='rbol BFS')
nx.draw(tree, with_labels=True)
```

```python
# El camino ms corto entre dos nodos
print(f"Camino ms corto entre 1 y 3: {nx.shortest_path(G, 1, 3)}")
# Lo representamos
plt.figure(5, figsize=(12, 12))
plt.subplot(221, title='Camino ms corto entre 1 y 3')
nx.draw(G, with_labels=True)
```
